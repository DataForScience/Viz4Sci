```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

## Generate some fake data

```python
t_max = 4 # signal duration (seconds)
sample_freq = 250 # points per second
N = t_max*sample_freq
t = np.linspace(0, t_max, N)
amp = np.array([1, .3, .1])
freq = np.array([1, 2, 10])
```

Combine all the components into the total signal

```python
total = np.zeros(1000)
components = []

n_freq = len(freq)

for i in range(n_freq):
    current = amp[i]*np.cos(2*np.pi*freq[i]*t)
    total += current
    
    components.append(current)
```

Get the color cycle

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

Vizualize in 3D

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("\n\nTime (s)", fontsize=24)
ax.set_ylabel("\n\nFrequency (Hz)", fontsize=24)
ax.set_zlabel("\n\nAmplitude", fontsize=24)

n_freq = np.max(freq)+2
x = np.linspace(0, 4, 1000)
y = np.ones(x.size)

# Plot the total signal
ax.plot(x, y*n_freq, total, linewidth=3, color=colors[1])

# Plot the amplitudes
z = np.zeros(n_freq*100)
z[freq*100] = amp

ax.plot(np.zeros(n_freq*100), np.linspace(0, n_freq, n_freq*100), z, linewidth=3, color=colors[3])

# Plot the components
y = np.ones(1000)
for i in range(0, len(components)):
    ax.plot(x, y*freq[i], components[i], linewidth=1.5, color=colors[0])

plt.setp(ax.get_xticklabels(), fontsize=18)    
plt.setp(ax.get_yticklabels(), fontsize=18)    
plt.setp(ax.get_zticklabels(), fontsize=18)    

ax.set_yticks(freq)
ax.set_yticklabels(freq)
ax.set_xlim(0, t_max)
ax.set_ylim(0, n_freq)
```
